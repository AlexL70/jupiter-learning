# Content-Based Filtering

## Read data

```python
import pandas as pd
movies = pd.read_csv("movies_data/movies.csv", sep=",")
movies
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer
tfidf = TfidfVectorizer(stop_words='english')
movies["overview"] = movies["overview"].fillna("") # Filling NaNs with empty string
tfidf_matrix = tfidf.fit_transform(movies["overview"])
pd.DataFrame(tfidf_matrix.toarray(), columns=tfidf.get_feature_names_out())
```

### Similarity Matrix

```python
from sklearn.metrics.pairwise import linear_kernel
similarity_matrix = linear_kernel(tfidf_matrix, tfidf_matrix)
similarity_matrix
```

### Find the most similar movies to the certain movie

```python
def similar_movies(movie_title: str, nr_movies: int) -> pd.DataFrame:
    # Get the index of the movie by its title
    movie_index = movies.loc[movies["title"] == movie_title].index[0]
    # Get the similarity scores of the movie with all other movies (including itself)
    # and enumerate them to keep track of the index
    scores = list(enumerate(similarity_matrix[movie_index].tolist()))
    # Remove the movie itself from the list (it will have a similarity score of 1)
    scores.pop(movie_index)
    # Sort the movies by similarity score in descending order
    scores.sort(reverse=True, key=lambda x: x[1])
    # Get the top n movies
    scores= scores[:nr_movies]
    return pd.DataFrame([movies.iloc[score[0]] for score in scores])

sm = similar_movies("Pulp Fiction", 10)
sm
```
